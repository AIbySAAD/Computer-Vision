```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('/content/th.jpg')

cvt_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(cvt_img,cmap='gray')
```

```python
histogram=cv2.calcHist([img],channels=[2],histSize=[256],ranges=[0,256],mask=None)
plt.plot(histogram) # this is actually telling how many times each pixel is happening
```

```python
hist_eq=cv2.equalizeHist(cvt_img)
plt.imshow(hist_eq,cmap='gray')
```
