```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('test image.jpg')
cvt_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
width_ratio=0.2
height_ratio=0.8
res_img=cv2.resize(cvt_img,(0,0),cvt_img,width_ratio,height_ratio)
plt.imshow(res_img)
```

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('test image.jpg')
cvt_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
f_img=cv2.flip(cvt_img,0)
plt.imshow(f_img)
```
