```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

black_image=np.zeros((512,512,3))
plt.imshow(black_image)
```

```python
rect_img=cv2.rectangle(black_image,(200,200),(300,300),thickness=3,color=(100,100,0))
plt.imshow(rect_img)
```

```python
rect_img=cv2.line(black_image,(500,300),(100,400),thickness=3,color=(100,100,0))
plt.imshow(rect_img)
```

```python
img=np.zeros((600,600,3))
```

```python
plt.imshow(img)
```

```python
cir_img=cv2.circle(img,radius=100,thickness=3,center=(200,200),color=(100,0,0))
plt.imshow(cir_img)
```

```python
cir_img=cv2.rectangle(img,pt1=(400,100),pt2=(500,300),thickness=3,color=(0,100,0))
plt.imshow(cir_img)
```
