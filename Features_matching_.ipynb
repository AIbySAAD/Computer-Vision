```python
import cv2
import matplotlib.pyplot as plt
book=cv2.imread('book.jpg')
book=cv2.cvtColor(book,cv2.COLOR_BGR2RGB)
book
```

```python
books=cv2.imread('books.jpg')
books=cv2.cvtColor(books,cv2.COLOR_BGR2RGB)
books
```

```python
def display(img):
  fig=plt.figure(figsize=(12,10))
  ax=fig.add_subplot(111)
  ax.imshow(img)
```

```python
orb=cv2.ORB_create() # creates an object can use to find detect and describe keypoints in image
```

```python
kp1,des1=orb.detectAndCompute(book,None) # No mask, returns keypoint and their description
```

```python
kp2,des2=orb.detectAndCompute(books,None) # No mask, returns keypoint and their description
```

```python
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches=bf.match(des1,des2)
```

```python
single_match=matches[0]
single_match.distance
```

```python
matches=sorted(matches,key=lambda x:x.distance)
```

```python
len(matches)
```

```python
matching_book=cv2.drawMatches(book,kp1,books,kp2,matches,None,flags=2) # no mask, and tell how it gonna draw the matches
```

```python
plt.imshow(matching_book)
```
