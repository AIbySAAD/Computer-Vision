```python
# Load modules and packages
#-----------------------------
# for template macthing
import cv2
# for displaying image
import matplotlib.pyplot as plt

# Load original Image
img=cv2.imread('Orange Tree.jpg')
# As matplotlib read image RGB unlike BGR
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Load Template
temp=cv2.imread('Orange Tree - temp.jpg')
# As matplotlib read image RGB unlike BGR
temp=cv2.cvtColor(temp,cv2.COLOR_BGR2RGB)
```

```python
# Show images
plt.imshow(img)
```

```python
plt.imshow(temp)
```

```python
# Lets match it
method=cv2.TM_CCOEFF_NORMED
matching=cv2.matchTemplate(img,temp,method)

plt.imshow(matching)
```

```python
# If you want to draw the square around that
#-------------------------------------------
minv,maxv,min_loc,max_loc=cv2.minMaxLoc(matching)
# width and height of template
width,height,channels=temp.shape
top_left=max_loc #(x,y)
bot_right=(top_left[0]+width,top_left[1]+height)
cv2.rectangle(img,top_left,bot_right,(0,255,0),3)

# Show Image
plt.imshow(img)
```
