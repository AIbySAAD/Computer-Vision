```python
import cv2
img=cv2.imread('dog.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img
```

```python


# Load the image
image = img

# Define the coordinates of the region of interest (ROI)
x1, y1 = 60, 60  # Top-left corner
x2, y2 = 120, 120  # Bottom-right corner

# Apply the mask using slicing
masked_image = image[y1:y2, x1:x2]

# Display the masked image
x1,y1=80,80
x2,y2=140,140
image[y1:y2, x1:x2]=masked_image
image
```
